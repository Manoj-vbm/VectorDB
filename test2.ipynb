```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = 'https://www.acquired.fm/episodes/microsoft'  # Replace with the actual URL

# Fetch the webpage content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

```

```python
print(soup.prettify())
```

```python
# Extract Episode Title
episode_title = soup.find('title').get_text(strip=True) if soup.find('title') else 'No Title Found'
print(episode_title)
```

```python
target_div = soup.find('div', class_='w-richtext')

# Find all paragraphs within this section
paragraphs = target_div.find_all('p')

# Combine the text from the paragraphs you are interested in
about = ' '.join([
    paragraphs[0].get_text(strip=True),
    paragraphs[1].get_text(strip=True)
])

# Print the combined extracted text
print(about)
```

```python
transcript_section = soup.find('div', class_='rich-text-block-6 w-richtext')
transcript = ''
if transcript_section:
    paragraphs = transcript_section.find_all('p')
    for p in paragraphs:
        transcript += p.get_text(strip=True) 
else:
    transcript = 'No Transcript Found'

print(transcript)
```

```python
# Extract Follow-Ups and Corrections
followups_section = soup.find('p', string='Follow-Ups and Corrections:')
followups_text = ''
if followups_section:
    followups_text += "Follow-Ups and Corrections:\n\n"
    ul = followups_section.find_next_sibling('ul')
    if ul:
        for li in ul.find_all('li'):
            followups_text += "• " + li.get_text(strip=True) + "\n\n"
else:
    followups_text = 'No Follow-Ups and Corrections Found'

print(followups_text)
```

```python
import requests
from bs4 import BeautifulSoup
import json

def extract_podcast_data(url):
    # Fetch the webpage content
    response = requests.get(url)
    html_content = response.content

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract Episode Title
    episode_title = soup.find('title').get_text(strip=True) if soup.find('title') else 'No Title Found'

    # Extract About section
    target_div = soup.find('div', class_='w-richtext')
    paragraphs = target_div.find_all('p') if target_div else []
    about = ' '.join([
        paragraphs[0].get_text(strip=True),
        paragraphs[1].get_text(strip=True)
    ]) if len(paragraphs) >= 2 else 'No About Section Found'

    # Extract Transcript
    transcript_section = soup.find('div', class_='rich-text-block-6 w-richtext')
    transcript = ''
    if transcript_section:
        paragraphs = transcript_section.find_all('p')
        for p in paragraphs:
            transcript += p.get_text(strip=True) + '\n\n'
    else:
        transcript = 'No Transcript Found'

    # Extract Follow-Ups and Corrections
    followups_section = soup.find('p', string='Follow-Ups and Corrections:')
    followups_text = ''
    if followups_section:
        followups_text += "Follow-Ups and Corrections:\n\n"
        ul = followups_section.find_next_sibling('ul')
        if ul:
            for li in ul.find_all('li'):
                followups_text += "• " + li.get_text(strip=True) + "\n\n"
    else:
        followups_text = 'No Follow-Ups and Corrections Found'

    # Create a dictionary with the extracted information
    podcast_data = {
        "title": episode_title,
        "about": about,
        "transcript": transcript,
        "followups_and_corrections": followups_text
    }

    return podcast_data
```

```python


# List of podcast episode URLs
episode_urls = [
    'https://www.acquired.fm/episodes/microsoft',
    'https://www.acquired.fm/episodes/renaissance-technologies',
    'https://www.acquired.fm/episodes/hermes',
    # Add more URLs here
]

all_podcast_data = []

```

```python
for url in episode_urls:
    try:
        episode_data = extract_podcast_data(url)
        all_podcast_data.append(episode_data)
        print(f"Processed: {episode_data['title']}")
    except Exception as e:
        print(f"Error processing {url}: {str(e)}")

# Save all data to a JSON file
with open('all_podcast_data.json', 'w', encoding='utf-8') as f:
    json.dump(all_podcast_data, f, ensure_ascii=False, indent=4)

print("All podcast data has been saved to 'all_podcast_data.json'")

# Print the JSON data (optional)
print(json.dumps(all_podcast_data, indent=4, ensure_ascii=False))
```

```python
# Add an 'id' field to each podcast entry
for i, podcast in enumerate(all_podcast_data, start=1):
    podcast['id'] = f'{i:02d}'  # This will create ids like '01', '02', '03', etc.
    podcast.update({'id': f'{i:02d}', **podcast})  # Add 'id' as the first entry

# Display the updated data
all_podcast_data
```

```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
# Prepare the data for Chroma
documents = [podcast['about'] for podcast in all_podcast_data]
ids = [podcast['id'] for podcast in all_podcast_data]
metadatas = [{'title': podcast['title'], 'url': podcast.get('url', '')} for podcast in all_podcast_data]

# Add the podcast data to the Chroma collection
collection.add(
    documents=documents,
    ids=ids,
    metadatas=metadatas
)

```

```python
# Example query to test the collection
query = "Tell me about dos"
results = collection.query(
    query_texts=[query],
    n_results=3
)

print(f"Query: {query}")
print("Results:")
for i, (id, distance) in enumerate(zip(results['ids'][0], results['distances'][0])):
    print(f"{i+1}. ID: {id}, Distance: {distance}")
    print(f"   Title: {next(podcast['title'] for podcast in all_podcast_data if podcast['id'] == id)}")
    print(f"   Snippet: {results['documents'][0][i][:200]}...")
```
